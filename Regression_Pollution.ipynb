```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
air_data0 = pd.read_excel('pb_half.xlsx').iloc[:1542]
air_data = air_data0.replace('None', -200, regex= True)
air_data.head()
# workbook_dataframe = pd.read_excel(workbook_filename).iloc[:10]
```

```python
air_data.shape
```

```python
air_data.dropna(axis=0, how='all')
```

```python
features = air_data
air_data.head()
```

```python
features = features.drop('To Date', axis=1)
features = features.drop('PM2.5', axis=1)
features = features.drop('NOx', axis=1)


# features = features.drop('From Date', axis=1)
# features = features.drop('From Date', axis=1)
# features = features.drop('From Date', axis=1)
# features = features.drop('From Date', axis=1)


print(features.head())
```

```python
labels = air_data['NOx'].values
```

```python
features = features.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3)
```

```python
print("X_trian shape --> {}".format(X_train.shape))
print("y_train shape --> {}".format(y_train.shape))
print("X_test shape --> {}".format(X_test.shape))
print("y_test shape --> {}".format(y_test.shape))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
# regressor.fit(X_train, y_train)
print(regressor.fit(X_train, y_train).coef_)
```

```python
print("Predicted values:", regressor.predict(X_test))
```

```python
print("R^2 score for liner regression: ", regressor.score(X_test, y_test))
```
